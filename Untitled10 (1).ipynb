```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
from pycaret.regression import setup, create_model, pull

# Load the California housing dataset
housing = fetch_california_housing(as_frame=True)
data = housing.frame

# Define the target column
target_column = 'MedHouseVal'

# Verify the target column
if target_column not in data.columns:
    raise ValueError(f"Target column '{target_column}' not found in the dataset.")

# PyCaret setup with updated parameters
reg_setup = setup(
    data=data,
    target=target_column,
    remove_outliers=True,            # Remove outliers
    outliers_threshold=0.05,         # Threshold for outlier removal
    normalize=True,                  # Enable normalization
    normalize_method='zscore',       # Normalization method
    pca=True,                        # Enable PCA
    pca_method='linear',             # PCA method
    train_size=0.7,                  # Train-test split ratio
    data_split_shuffle=False,        # Do not shuffle data during the split
    verbose=False                    # Suppress detailed logs
)

# List of regression models to evaluate
models = [
    'et', 'rf', 'lightgbm', 'gbr', 'xgboost', 'huber', 'br', 'dt',
    'ridge', 'lar', 'lr', 'en', 'ada', 'knn', 'lasso', 'llar', 'par', 'omp', 'dummy'
]

# Dictionary to store results
results = {}

# Loop through each model
for model in models:
    print(f"Training model: {model}")
    trained_model = create_model(model)  # Train the model
    metrics = pull()  # Retrieve metrics from the latest run

    # Debugging: Print available columns and data for the first model
    if model == models[0]:
        print("\nAvailable Metrics Columns:", metrics.columns.tolist())
        print("\nMetrics Dataframe (Sample):\n", metrics.head())

    # Retrieve the R² value from metrics
    if 'R2' in metrics.columns:  # Check if 'R2' is a column
        r2_value = metrics['R2'].iloc[0]  # Retrieve the first row's 'R2' value
        results[model] = r2_value
    else:
        print(f"Warning: 'R2' not found for model {model}. Skipping.")

# Find the best model based on R²
if results:
    best_model_name = max(results, key=results.get)
    print(f"\nBest Model: {best_model_name} with R2 = {results[best_model_name]}")
else:
    print("\nNo models had a valid R2 metric.")


```

```python
from pycaret.regression import setup, create_model, plot_model
from sklearn.datasets import fetch_california_housing

# Load the California housing dataset
housing = fetch_california_housing(as_frame=True)
data = housing.frame

# Define the target column
target_column = 'MedHouseVal'

# PyCaret setup
setup(
    data=data,
    target=target_column,
    normalize=True,
    verbose=False
)

# Train a Random Forest model
rf = create_model('rf', n_jobs=1)  # Disable parallelism for model creation

# Plot error and learning curve without parallel processing
plot_model(rf, plot='error')
plot_model(rf, plot='learning')
```
